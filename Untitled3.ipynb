```python
import os
import re
from xml.etree import ElementTree
from docx import Document
from google.colab import files

# Namespaces XML do Microsoft Word necessários para interceptar o Track Changes
WORD_NAMESPACE = "{http://openxmlformats.org}"
TAG_TEXTO = WORD_NAMESPACE + "t"
TAG_DELECAO = WORD_NAMESPACE + "del"


# ==============================================================================
# REQUISITO 2: ENGINE PARA ACEITAR AS MARCAS DE ALTERAÇÃO (TRACK CHANGES)
# ==============================================================================
def obter_texto_com_revisoes_aceitas(elemento_docx):
    """Varre a estrutura XML do Word e aceita as revisões virtualmente.

    Ignora trechos deletados (<w:del>) e mantém apenas os textos válidos (<w:t>).
    """
    xml_str = elemento_docx._element.xml

    # Se o parágrafo ou célula não tiver revisões, retorna o texto comum direto
    if "w:del" not in xml_str and "w:ins" not in xml_str:
        return elemento_docx.text

    # Caso tenha revisões, reconstrói o texto interpretando as tags XML
    arvore = ElementTree.fromstring(xml_str)
    fragmentos_texto = []

    for no in arvore.iter():
        if no.tag == TAG_DELECAO:
            continue  # Requisito 2: Pula e descarta elementos deletados
        if no.tag == TAG_TEXTO and no.text:
            fragmentos_texto.append(no.text)  # Mantém o texto inserido/final

    return "".join(fragmentos_texto)


# ==============================================================================
# REQUISITO 4: EXTRAÇÃO SEQUENCIAL DE NÚMEROS (TEXTO + TABELAS)
# ==============================================================================
def extrair_e_normalizar_numeros(caminho_arquivo, idioma):
    """Acessa parágrafos e tabelas sequencialmente, limpa as pontuações e

    converte cada número para o padrão numérico float do Python.
    """
    doc = Document(caminho_arquivo)
    numeros_normalizados = []
    detalhes_originais = []

    # Expressão regular para capturar números inteiros e decimais
    regex_numeros = r"\b\d+(?:[\.,]\d+)*\b"

    def processar_elemento(elemento_docx, localizacao_str):
        """Aplica a limpeza de revisão e a regex de números em cada bloco de

        texto.
        """
        texto_limpo = obter_texto_com_revisoes_aceitas(elemento_docx)

        for match in re.finditer(regex_numeros, texto_limpo):
            num_original = match.group()

            # Requisito 3: Padronização inteligente dependendo do idioma
            if idioma == "en":
                # Inglês: Remove vírgula de milhar (ex: 1,500.50 -> 1500.50)
                num_limpo = num_original.replace(",", "")
            elif idioma == "pt":
                # Português: Remove ponto de milhar e converte vírgula para ponto (ex: 1.500,50 -> 1500.50)
                num_limpo = num_original.replace(".", "").replace(",", ".")

            try:
                # Converte para float permitindo a comparação matemática perfeita (ex: 12 == 12.0)
                valor_float = float(num_limpo)
                numeros_normalizados.append(valor_float)

                # Gera um trecho do contexto ao redor do número para facilitar a busca manual no Word
                contexto = texto_limpo[
                    max(0, match.start() - 15) : min(
                        len(texto_limpo), match.end() + 15
                    )
                ].strip()

                detalhes_originais.append(
                    {
                        "original": num_original,
                        "local": localizacao_str,
                        "contexto": f"... {contexto} ...",
                    }
                )
            except ValueError:
                continue

    # Varredura Sequencial Parte A: Corpo do Texto
    for num_p, paragrafo in enumerate(doc.paragraphs, 1):
        processar_elemento(paragrafo, f"Parágrafo {num_p}")

    # Varredura Sequencial Parte B: Tabelas (Requisito 4)
    for num_t, tabela in enumerate(doc.tables, 1):
        for num_r, linha in enumerate(tabela.rows, 1):
            for num_c, celula in enumerate(linha.cells, 1):
                for paragrafo_celula in celula.paragraphs:
                    local_str = f"Tabela {num_t}, Linha {num_r}, Célula {num_c}"
                    processar_elemento(paragrafo_celula, local_str)

    return numeros_normalizados, detalhes_originais


# ==============================================================================
# REQUISITO 3: ALGORITMO DE COMPARAÇÃO E RELATÓRIO DE DISCREPÂNCIAS
# ==============================================================================
def comparar_documentos(caminho_en, caminho_pt):
    print("\n" + "=" * 70)
    print("🤖 INICIANDO COMPARAÇÃO SEQUENCIAL DE NÚMEROS")
    print("=" * 70)

    # Processa os dois arquivos gerando as listas numéricas normalizadas
    nums_en, detalhes_en = extrair_e_normalizar_numeros(caminho_en, "en")
    nums_pt, detalhes_pt = extrair_e_normalizar_numeros(caminho_pt, "pt")

    print(f"-> Total de números mapeados no arquivo EN: {len(nums_en)}")
    print(f"-> Total de números mapeados no arquivo PT: {len(nums_pt)}\n")

    erros_encontrados = 0
    total_comparacoes = max(len(nums_en), len(nums_pt))

    # Varre as duas listas de números elemento por elemento
    for i in range(total_comparacoes):

        # Caso 1: O arquivo em Inglês tem mais números na sequência que o de Português
        if i >= len(nums_pt):
            print(f"❌ ERRO [Posição Sequencial {i+1}]: Número extra sobrou no arquivo INGLÊS!")
            print(f"   Localização: {detalhes_en[i]['local']}")
            print(f"   Valor encontrado: '{detalhes_en[i]['original']}'")
            print(f"   Contexto: {detalhes_en[i]['contexto']}\n" + "-" * 50)
            erros_encontrados += 1
            continue

        # Caso 2: O arquivo em Português tem mais números na sequência que o de Inglês
        if i >= len(nums_en):
            print(f"❌ ERRO [Posição Sequencial {i+1}]: Número extra sobrou no arquivo PORTUGUÊS!")
            print(f"   Localização: {detalhes_pt[i]['local']}")
            print(f"   Valor encontrado: '{detalhes_pt[i]['original']}'")
            print(f"   Contexto: {detalhes_pt[i]['contexto']}\n" + "-" * 50)
            erros_encontrados += 1
            continue

        # Caso 3: Ambos possuem número na mesma posição, mas os valores são matematicamente divergentes
        if nums_en[i] != nums_pt[i]:
            print(f"❌ ERRO DE DIVERGÊNCIA na posição sequencial {i+1}:")
            print(f"   • INGLÊS:   '{detalhes_en[i]['original']}' | {detalhes_en[i]['local']} | Contexto: {detalhes_en[i]['contexto']}")
            print(f"   • PORTUGUÊS: '{detalhes_pt[i]['original']}' | {detalhes_pt[i]['local']} | Contexto: {detalhes_pt[i]['contexto']}")
            print("-" * 70)
            erros_encontrados += 1

    print("\n" + "=" * 70)
    if erros_encontrados == 0:
        print("✅ SUCESSO! Todos os números batem perfeitamente na ordem em ambos os arquivos.")
    else:
        print(f"FIM DA ANÁLISE. Foram encontradas {erros_encontrados} discrepâncias no total.")
    print("=" * 70)


# ==============================================================================
# REQUISITO 1: MÓDULO DE INTERFACE E UPLOAD NO COLAB
# ==============================================================================
print("1) Selecione o arquivo de Word em INGLÊS:")
upload_en = files.upload()  # Abre o seletor visual do navegador no Colab
nome_arquivo_en = list(upload_en.keys())[0]  # Armazena o nome/caminho gerado

print("\n2) Selecione o arquivo de Word em PORTUGUÊS:")
upload_pt = files.upload()  # Abre o seletor visual para o segundo arquivo
nome_arquivo_pt = list(upload_pt.keys())[0]  # Armazena o nome/caminho gerado

# Dispara a checagem passando as variáveis criadas no Upload
comparar_documentos(nome_arquivo_en, nome_arquivo_pt)
```
